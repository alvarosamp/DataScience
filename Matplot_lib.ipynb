##Matplot example workflow

```python
# 0. Import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Simple  Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

### Which one should you use ?(pyplot vs matplotlib 00 method)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method


```python
heart.head()
```

```python
over_50 = heart[heart["age"] >  50]
len(over_50)
```

```python
over_50.head()
```

```python
#pyplot method
over_50.plot(kind = 'scatter',
             x='age',
             y ='chol',
             c = 'target')
```

```python
#OO method
fix,ax = plt.subplots(figsize = (10,6))
over_50.plot ( kind = 'scatter',
              x = 'age',
               y= 'chol',
               c = 'target',
               ax = ax)
ax.set_xlim([40,100])
```

### // 6

```python
## OO method from scrtatch
fig, ax = plt.subplots(figsize = (10,6))

#Plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50['target'])

# Customize the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(),title = "Target");

#Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = '--');
```

### //7

```python
  #Subplot of chol,age , thalach
fig, (ax0,ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10),
                              sharex = True)
 #add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                       y = over_50["chol"],
                       c = over_50["target"])

# Customize ax0
ax0.set(title = "Heart Disease and Cholesterol Levels",
        xlabel = "Age",
        ylabel = "Cholesterol")

#Add a meanline
ax0.axhline (y = over_50["chol"].mean(),
             linestyle ="--") ;

#Add data to ax1
scatter = ax1.scatter( x = over_50["age"],
                      y = over_50["thalach"],
                       c = over_50["target"])
#Customize ax1
ax1.set( title = "Heart Disease and Max Heart Rate",
        xlabel = "Age",
         ylabel = "Max Heart Rate");

#Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title ="Target")

#Add a meanline
ax1.axhline( y = over_50["thalach"].mean(),
            linestyle ="--");

#Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize = 16,fontweight = "bold")

```
